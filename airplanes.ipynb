```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import seaborn as sns
%matplotlib inline
airplane=sns.load_dataset('flights')
car = sns.load_dataset('car_crashes')
```

```python
sns.catplot(x='passengers',y='month',jitter=False,data=airplane)
#Very colorfu and easy to tell apart, but hard to visually see
```

```python
sns.catplot(x='passengers',y='month',kind='box',data=airplane)
#Much better in telling the data, not cramped, easy to read
```

```python
sns.catplot(x="passengers", y="month", kind="boxen",
            data=airplane);
#Much more flare than the previous one, but not neccesarry, depends on the data and story
```

```python
sns.catplot(x="passengers", y="month", kind="violin",
            data=airplane);
#Very good for noting density, good for showing repetative values
```

```python
sns.catplot(x="passengers", y="month", kind="bar",
            data=airplane);
#very simple to understand, but the could be cut in some form to only show the top
```

```python
sns.relplot(x='year',y='passengers',size="month",data=airplane)
#Very dumb for a timewise function, would be good in a more random scatterplot
```

```python
sns.relplot(x='year',y='passengers',hue='month',data=airplane)
#good at giving additional information, but again not good for timewise
```

```python
g = sns.relplot(x="year", y="passengers", kind="line", data=airplane)
#Much more streamlined, to tell on increasing passengers
```

```python
sns.catplot(x="year", y="passengers", hue="month", kind="point", data=airplane);
#Very nice for noting the different catagorical figures on timewise
```
